```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random

random.seed(0)

x = np.random.randn(50)
sigma = np.random.randn(len(x))/3
y = x**3-(1/2)*x**2-6*x+1 + sigma
plt.scatter(x,y)
plt.show()

data = pd.DataFrame(np.array([x,y]).T, columns=["x", "y"])
data.to_csv("Datos_Ajuste_Pablo_Calatayud.csv")
```
